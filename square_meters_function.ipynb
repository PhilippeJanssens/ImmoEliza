```python
import pandas as pd
import numpy as np

df_ElizaDATASET = pd.read_csv("fullDataset.csv")
#print(df_ElizaDATASET)
df_eliza = pd.read_csv("eliza.csv")
#print(df_eliza.columns.values)
#print(type(df_eliza[["Terrain"]]))
df_ImmoWeb = pd.read_csv("immoweb_batch_1_5_copy.csv")


def square_meters_cleaning(df,columns):
    """
    Function converting the square meters data into floats
    : df: DataFrame to clean
    : columns: list of column labels for which the data needs cleaning
    : return: cleaned DataFrame
    """
    for column in columns:
        for index,element in df[[column]].iterrows():
            if isinstance(element[column],str) and "m2" in element[column]:
                df = df.replace(element[column],float(element[column][:-3]))
        return df

def square_meters_cleaning_ImmoWeb(df,columns):
    for column in columns:
        for index,element in df[[column]].iterrows():
            if isinstance(element[column],str) and "m²" in element[column]:
                df = df.replace(element[column],float(element[column][:-3]))
        return df
    
def price_ImmoWeb(df,column):
    for index,element in df[[column]].iterrows():
        element_to_clean = element[column]
        
        if isinstance(element_to_clean,str) and"€" in element_to_clean:
            element_to_clean = element_to_clean[: -2]
            while "." in element_to_clean:
                splitted_number = element_to_clean.split(".")
                element_to_clean = "".join(splitted_number)
            if "Faire offre à partir de" in element_to_clean:
                element_to_clean = element_to_clean.replace("Faire offre à partir de ","")
            print(element_to_clean)
            df = df.replace(element[column],int(element_to_clean))
    for index,element in df[[column]].iterrows():
        if element[column] == "Oui" or element[column] == "Non":
            df = df.replace(element[column],"NaN")
    return df
    
print(price_ImmoWeb(df_ImmoWeb,"Prix")["Prix"])
```
