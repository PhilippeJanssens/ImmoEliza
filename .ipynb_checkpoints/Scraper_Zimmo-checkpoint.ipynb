```python
driver = webdriver.Chrome()
```

```python
class Zimmo():
    """
    This class and it's scrape method can be used to gather all data from the website
    www.zimmo.be.
    """
    def __init__(self, start_url="https://www.zimmo.be/fr/biens/?status=1&hash=98b01d95708f84f669bdb6c66c482614&priceIncludeUnknown=1&priceChangedOnly=0&bedroomsIncludeUnknown=1&bathroomsIncludeUnknown=1&constructionIncludeUnknown=1&livingAreaIncludeUnknown=1&landAreaIncludeUnknown=1&commercialAreaIncludeUnknown=1&yearOfConstructionIncludeUnknown=1&epcIncludeUnknown=1&queryCondition=and&includeNoPhotos=1&includeNoAddress=1&onlyRecent=0&onlyRecentlyUpdated=0&isPlus=0&excludedEstates%5B0%5D=JWV12&excludedEstates%5B1%5D=JX5TS&excludedEstates%5B2%5D=JX571&excludedEstates%5B3%5D=JXA9T&region=none&pagina=1#gallery"):
        self.start_url = start_url
        self.advert_urls = []
        self.advert_details = []
        
    def page_advert_urls(self, page_url, url_piece = "a-vendre"):
        """
        Method that gathers urls for each individual ad. 
        
        :page_url: a url that links to a page containing several urls of individual ads
        :url_piece: search string that should be contained in the links for it to be an ad.
        """
        
        driver.get(page_url)
        time.sleep(random.uniform(1.0, 2.0))
        for elem in driver.find_elements_by_tag_name("a"):
            link = elem.get_attribute("href")
            if not link is None and url_piece in link:
                self.advert_urls.append(link)

    def readPagination(self):
        """
        readPagination gathers all pages that contain independent ads. The first one is hardcoded and 
        aferwards it formats the URL as a string until no new ads are found.
        """
        # initialize a list for storing all the seperate links to individual ads.
        #all_advert_urls = []
        #advert_urls = []
        counter = 2
        print(self.start_url)
        self.page_advert_urls(self.start_url)
        
        previous_adverts_amount = 0
        adverts_amount = len(self.advert_urls)

        while previous_adverts_amount < adverts_amount:
            previous_adverts_amount = adverts_amount
            
            page_url = f"https://www.zimmo.be/fr/biens/?status=1&hash=98b01d95708f84f669bdb6c66c482614&priceIncludeUnknown=1&priceChangedOnly=0&bedroomsIncludeUnknown=1&bathroomsIncludeUnknown=1&constructionIncludeUnknown=1&livingAreaIncludeUnknown=1&landAreaIncludeUnknown=1&commercialAreaIncludeUnknown=1&yearOfConstructionIncludeUnknown=1&epcIncludeUnknown=1&queryCondition=and&includeNoPhotos=1&includeNoAddress=1&onlyRecent=0&onlyRecentlyUpdated=0&isPlus=0&excludedEstates%5B0%5D=JWV12&excludedEstates%5B1%5D=JX5TS&excludedEstates%5B2%5D=JX571&excludedEstates%5B3%5D=JXA9T&region=none&pagina={counter}#gallery"
            self.page_advert_urls(page_url)
            
            adverts_amount = len(self.advert_urls)
            counter += 1
            
    def readAdPage(self):
        """
        Method the reads all the information on the page and writes it to a dict using the key, 
        value pairs used on the website in question. 
    
        :inp_url: get the input URL that we will be gathering information from.
        :return: a dict that stores scraped data
        """
        for advert_url in ["https://www.zimmo.be/fr/poperinge-8970/a-vendre/maison/JVC2I/?search=98b01d95708f84f669bdb6c66c482614&boosted=1"]:#self.advert_urls[:1]:
            # initialize some variables.
            detailsKeys = []
            detailsValues = []
            details = {}
    
            # grab the URL page code
            website = driver.get(advert_url)
            time.sleep(random.uniform(4.0, 5.0))
            print(type(website))
    
            # find the tables in the page and split data in 2 types, keys and values.
            for desc_list in driver.find_elements_by_xpath("/html/body/div/div/section/div/div/section/div/div/div/ul/li[@class='show-on-print']"):
                detailsKeys.append(desc_list.find_element_by_tag_name("strong").get_attribute("innerText"))
                detailsValues.append(desc_list.find_element_by_tag_name("span").get_attribute("innerText"))
             
            for desc_list in driver.find_elements_by_tag_name("li"):#by_xpath("/html/body/div/div/section/div/div/section/div/div/div/ul/li"):
                print("Suivant", desc_list.get_attribute("class"),desc_list)
                #time.sleep(random.uniform(0.5, 1))
                for key in desc_list.find_elements_by_tag_name("strong"):
                    detailsKeys.append(key.text)
                for value in desc_list.find_elements_by_tag_name("span"):
                    detailsValues.append(value.text)
            
            for desc_list in driver.find_elements_by_class_name("feature-label"):
                detailsKeys.append(desc_list.text)
                
            for desc_list in driver.find_elements_by_class_name("col-xsm-4 info-name"):
                detailsKeys.append(desc_list.text)
            
            for desc_list in driver.find_elements_by_class_name("feature-value"):
                detailsValues.append(desc_list)
                
            for desc_list in driver.find_elements_by_class_name("col-xsm-8 info-value"):
                detailsValues.append(desc_list)
            
            # build a dict containing all gathered data
            for idx, x in enumerate(detailsKeys):
                details[x] = detailsValues[idx]
                
            self.advert_details.append(details)
```
