```python
import pandas as pd

df_ElizaDATASET = pd.read_csv("fullDataset.csv")
#print(df_ElizaDATASET)
df_eliza = pd.read_csv("eliza.csv")
#print(df_eliza.columns.values)


```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)

# Get data from "df_eliza" into the right columns of "df_ElizaDATASET"
df_ElizaDATASET["Source"] = "Elissa"
df_ElizaDATASET["Locality"] = df_eliza["Ville"]
df_ElizaDATASET["Type of property"] = df_eliza["Type du bien"]
df_ElizaDATASET["Subtype of property"] = df_eliza["Nan"]
## Use .fillna to Replace NULL(or NaN?) values in column with the values from other column:
df_ElizaDATASET["Price"] = df_eliza["Unnamed: 1"].fillna(df_eliza["Prix"])
df_ElizaDATASET["Type of sale"] = "NaN"
df_ElizaDATASET["Number of rooms"] = df_eliza["Chambres"]
df_ElizaDATASET["Area"] = "NaN"
df_ElizaDATASET["Fully equipped kitchen"] = "NaN"
df_ElizaDATASET["Furnished"] = "NaN"
df_ElizaDATASET["Open fire"] = "NaN"

df_ElizaDATASET["Terrace"] = df_eliza["Terrasse"]
df_ElizaDATASET["Garden"] = df_eliza["Jardin"]
df_ElizaDATASET["Surface of the land"] = df_eliza["Terrain"]
df_ElizaDATASET["Surface area of the plot of land"] = "NaN"
df_ElizaDATASET["Number of facades"] = "NaN"
df_ElizaDATASET["Swimming pool"] = "NaN"
df_ElizaDATASET["State of the building"] = "NaN"
df_ElizaDATASET["Source"] = "Elissa"
df_ElizaDATASET
```

```python

# translate all..in column "Type of property"
df_ElizaDATASET["Type of property"] = df_ElizaDATASET["Type of property"].replace("Maison","House").replace("Appartement", "Apartment").replace("Commerce", "Business").replace("Immeuble de rapport", "Rental property")

df_ElizaDATASET
```

```python
# change column "Terrace" values into Yes/No
for index, row in df_ElizaDATASET.iterrows():
    if type(row["Terrace"]) == str:
        continue
    else:
        ifor_val = 0
    df_ElizaDATASET.at[index, "Terrace"] = ifor_val

df_ElizaDATASET
```

```python
# change column "Garden" values into Yes/No
for index, row in df_ElizaDATASET.iterrows():
    if type(row["Garden"]) == str:
        ifor_val = 1
    else:
        ifor_val = 0
    df_ElizaDATASET.at[index, "Garden"] = ifor_val

df_ElizaDATASET
```
